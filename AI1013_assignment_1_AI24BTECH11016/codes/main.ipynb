For two columns with firing rates 
\begin{align*}
    \lambda_i = [\lambda_i(1) \dots \lambda_i(N)] \\
    \lambda_j = [\lambda_j(1) \dots \lambda_j(N)]
\end{align*}
The, relative entropy distance $D_{ij}$ per neuron is defined as
$$
    D_{ij} = \frac{1}{N} \sum_{n1}^{N} \lambda_i(n)\log(\frac{\lambda_i(n)}{\lambda_j(n)})- \lambda_i(n) + \lambda_j(n)
$$

```python
relative_entropy_distances = []

#function to calculate length of the array since 
def length(arr):
    return np.sum(~pd.isna(arr))

#function to calculate relative entropy distance between two columns without laplace smoothing
def rel_entropy_dist(arr1, arr2):
    D_ij = 0
    if length(arr1)==length(arr2):
        for lambda_ in range(length(arr1)):
            length_final = length(arr1)
            if arr1.iloc[lambda_]==0 or arr2.iloc[lambda_]==0:
                D_ij += 0
                length_final -= 1
            else:
                D_ij += arr1.iloc[lambda_]*np.log(arr1.iloc[lambda_]/arr2.iloc[lambda_]) - arr1.iloc[lambda_] + arr2.iloc[lambda_]
        D_ij/=length_final
        #print(length(arr1))
    return D_ij


#calculating D_ij and D_ji for the columns i and j and appending it to the relative entropy array
for i in range(0, 18, 2):
    #D_ij
    relative_entropy_distances.append(rel_entropy_dist(firing_rates.iloc[2:, i], firing_rates.iloc[2:, i+1]))
    #D_ji
    relative_entropy_distances.append(rel_entropy_dist(firing_rates.iloc[2:, i+1], firing_rates.iloc[2:, i]))


#printing relative entropy distances    
relative_entropy_distances = np.array(relative_entropy_distances, dtype = float)
relative_entropy_distances
```

```python

relative_entropy_distances_laplace = []

#calculating relative entropy distance between two columns using laplace smoothing
def rel_entropy_dist_laplace(arr1, arr2):
    D_ij = 0
    if length(arr1)==length(arr2):
        for lambda_ in range(length(arr1)):
            length_final = length(arr1)
            if arr1.iloc[lambda_]==0 and arr2.iloc[lambda_]==0:
                arr1.iloc[lambda_] = 1e-10
                arr2.iloc[lambda_] = 1e-10
            if arr1.iloc[lambda_]==0 and arr2.iloc[lambda_]!=0:
                arr1.iloc[lambda_] = 1e-10
            if arr1.iloc[lambda_]!=0 and arr2.iloc[lambda_]==0:
                arr2.iloc[lambda_] = 1e-10
            D_ij += arr1.iloc[lambda_]*np.log(arr1.iloc[lambda_]/arr2.iloc[lambda_]) - arr1.iloc[lambda_] + arr2.iloc[lambda_]
        D_ij/=length(arr1)
        #print(length(arr1))
    return D_ij


#calculating D_ij and D_ji for the columns i and j and appending it to the relative entropy array
for i in range(0, 18, 2):
    #D_ij
    relative_entropy_distances_laplace.append(rel_entropy_dist_laplace(firing_rates.iloc[2:, i], firing_rates.iloc[2:, i+1]))
    #D_ji
    relative_entropy_distances_laplace.append(rel_entropy_dist_laplace(firing_rates.iloc[2:, i+1], firing_rates.iloc[2:, i]))


#printing relative entropy distances    
relative_entropy_distances_laplace = np.array(relative_entropy_distances_laplace, dtype = float)
relative_entropy_distances_laplace
```

For the two columns, finding $L_1$ distances per neuron is given as
$$
    L_{ij} = \frac{1}{N} \sum_{n=1}^{N} |\lambda_i(n) - \lambda_j(n)|
$$

```python
L_1_distances = []
#function to calculate l_1 distances between two arrays
def l1_dist(arr1, arr2):
    L_ij = 0
    if length(arr1) == length(arr2):
        for lambda_ in range(length(arr1)):
            L_ij += abs(arr1.iloc[lambda_]-arr2.iloc[lambda_])
        L_ij /= length(arr1)
    return L_ij

#appending the L_1 distances between each columns to the L_1 distances array
for i in range(0, 18, 2):
        L_1_distances.append(l1_dist(firing_rates.iloc[2:, i], firing_rates.iloc[2:, i+1]))
        L_1_distances.append(l1_dist(firing_rates.iloc[2:, i+1], firing_rates.iloc[2:, i]))
        
        
#printing L_1 distances
L_1_distances =  np.array(L_1_distances, dtype = float)
L_1_distances
```

Using the method of least squares to find the line of best fit in the form $y = mx$, 
$$
    m = \frac{\sum_{i=1}^n x_iy_i}{\sum_{i=1}^n x_i^{2}}
$$

```python
#using least squares method to find the slope of best fit
def lsq(x_vals, y_vals):
    numerator = 0
    denominator = 0
    for i in range(len(x_vals)):
        numerator += x_vals[i]*y_vals[i]
        denominator += x_vals[i]**2
    m = numerator/denominator
    return m
```

```python
import matplotlib.pyplot as plt

#plotting relative entropy distances against the inverse of the average search delays
x_vals = relative_entropy_distances
y_vals = 1/average_search_delays

#finding the best fit line for the data points
slope_rel_entropy_distances = lsq(x_vals, y_vals)
y_fit = slope_rel_entropy_distances*x_vals #(y = mx)


#labels and legends
plt.plot(x_vals, y_fit, color = "blue")
plt.scatter(x_vals, y_vals, color = "red")
plt.xlabel("Relative Entropy Distances")
plt.ylabel("Inverse of Search Delays")
plt.title("Without laplace smoothing")

plt.savefig("../figs/rel_entropy_dist_vs_inv_search_delays_withoutSmoothing")

#printing the slope found using least squares
print(f"Slope of line of best fit for relative entropy distances = {slope_rel_entropy_distances}")
```

```python
#plotting relative entropy distances calculated using laplace smoothing against the inverse of the average search delays
x_vals_laplace = relative_entropy_distances_laplace
y_vals_laplace = 1/average_search_delays

#finding the best fit line for the data points
slope_rel_entropy_distances_laplace = lsq(x_vals_laplace, y_vals_laplace)
y_fit_laplace = slope_rel_entropy_distances_laplace*x_vals_laplace #(y = mx)


#labels and legends
plt.plot(x_vals_laplace, y_fit_laplace, color = "blue")
plt.scatter(x_vals_laplace, y_vals_laplace, color = "red")
plt.xlabel("Relative Entropy Distances")
plt.ylabel("Inverse of Search Delays")
plt.title("With laplace smoothing")
plt.show()
plt.savefig("../figs/rel_entropy_dist_vs_inv_search_delays_withSmoothing")


#printing the slope found using least squares
print(f"Slope of line of best fit for relative entropy distances with laplace smoothing = {slope_rel_entropy_distances_laplace}")
```

```python
#plotting L_1 distances against the inverse of the search delays
x_lvals = L_1_distances
y_lvals = 1/average_search_delays

#finding the best slope for the data points using least squares
slope_l1_distances = lsq(x_lvals, y_lvals) 
y_lfit = slope_l1_distances*x_lvals

#labels and legends
plt.scatter(x_lvals, y_lvals, color = "red")
plt.plot(x_lvals, y_lfit, color = "blue")
plt.xlabel("$l_1$ distances")
plt.ylabel("Inverse of Search Delays")
plt.show()
plt.savefig("../figs/l1_dist_vs_inv_search_delays")

#printing the slope of the line of best fit
print(f"Slope of line of best fit for $l_1$ distances is {slope_l1_distances}")



```

Using MSE(Mean Squared Error) to determine line of best fit
$$
    MSE = \frac{1}{N}\sum_{i=1}^{N}(\hat{y_i} - y_i)^2
$$

```python
#calculating mean squared error for both the fits
mse_l1_distances = np.mean((y_lfit-y_lvals)**2)
mse_re_distances = np.mean((y_fit-y_vals)**2)
mse_re_distances_laplace = np.mean((y_fit_laplace - y_vals_laplace)**2)

#printing the mean squared error
print(f"mse for $l_1$ distances is {mse_l1_distances}")
print(f"mse for relative entropy distances without laplace smoothing is {mse_re_distances}")
print(f"mse for relative entropy distances with laplace smoothing is {mse_re_distances_laplace}")
```

Hence, the relative entropy distances line is a better fit since the MSE(Mean Squared Error) is lower, we can also see this graphically <br>
Also, we get a better fit for the relative entropy distances when we use laplace smoothing

We clearly see that the relative entropy distances have a strong linear correlation with the inverse of the search delays. This is because, when an oddball image is more distinct (higher relative entropy), the neural responses diverge more from the distractor, leading to faster detection (lower search delay).

```python
#plotting the mean against the standard deviation for the first random half
y = mean_search_delays
x = std_search_delays

#finding the line of best fit line for the data points to find the shape parameter a
slope_mean_vs_std = lsq(std_search_delays, mean_search_delays)
y_fit_mean_vs_std = slope_mean_vs_std*x

#labels and legends
plt.plot(x, y_fit_mean_vs_std)
plt.scatter(x, y, color = "red")
plt.xlabel("Standard Deviation of Search Delays")
plt.ylabel("Mean Search Delays")
plt.savefig("../figs/standard_dev_vs_mean_search_delays")


#printing shape parameter
shape_parameter = slope_mean_vs_std**2
print(f"The shape parameter is {shape_parameter}")
```


By plotting $\mu$ vs $\sigma$ we get the slope of best fit $\frac{\mu}{\sigma}$ as the square root of the shape parameter $a$

```python
#plotting the mean search delays against the square of the standard deviation of the search delays for the first half
x_1 = np.square(std_search_delays)
y_1 = mean_search_delays

#finding the line of best fit using least squares
slope_mean_vs_std_sq = lsq(x_1, y_1)
y_1_fit = slope_mean_vs_std_sq*x_1


#labels and legends
plt.scatter(x_1, y_1, color = "red")
plt.plot(x_1, y_1_fit, color = "blue")
plt.xlabel(r"$\sigma^{2}$(Variance of Search Delays)")
plt.ylabel(r"$\mu$(Mean of Search Delays)")
plt.savefig("../figs/variation_vs_mean_search_delays")
#printing the rate parameter
rate_parameter = slope_mean_vs_std_sq
print(f"Rate Parameter is {rate_parameter}")#b
```

By plotting $\mu$ vs $\sigma^2$ we get the slope of best fit as $\frac{\mu}{\sigma^2}$ which is the rate parameter $b$

The Kolmogorov-Smirnov test statistic determines the infinity norm between the two functions <br>
Kolmogorov-Smirnov test statistic (KS) can be estimated as
$$
    KS(F, G) = max|F(x) - G(x)|
$$

```python
#function to calculate the kolmogorov test statistic
def calc_ks(x_arr, y_arr):
    max_val = -np.inf
    for i in range(len(x_arr)):
        if abs(y_arr[i]-x_arr[i])>max_val:
            max_val = abs(y_arr[i]-x_arr[i])
            
    return max_val
        
```

```python
#calculating the kolmogorov test statistic for the gamma cdf and the empirical cdf
kolmogorov_smirnov_test_statistic = calc_ks(gamma_cdf, ys)
kolmogorov_smirnov_test_statistic
```
